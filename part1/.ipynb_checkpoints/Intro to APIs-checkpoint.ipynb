```python
series_set = ['SP500', 
              'DJIA', 
              'NASDAQCOM',
              'DGS1', 
              'DGS10', 
              'DGS30',
              'DCOILWTICO',
              'DEXUSEU', 
              'GOLDAMGBD228NLBM', 
              'GDP'] 
```

# Linear Regression Challenge

```python
daily_changes = df_join.pct_change()[1:]
```

```python
import statsmodels.api as sm
```

```python
x = daily_changes['SP500']
y = daily_changes['XOM']
x = sm.add_constant(x)
model = sm.OLS(y, x)
results = model.fit()
results.summary()
```

```python
results.params
```

```python
results.rsquared
```

# Comparing All Variables

```python
for column in daily_changes.columns:
    x = daily_changes[column]
    y = daily_changes['XOM']
    x = sm.add_constant(x)
    model = sm.OLS(y, x)
    results = model.fit()
    print(column, results.rsquared)
```

```python
rsquared_data = {}

for column in daily_changes.columns:
    x = daily_changes[column]
    y = daily_changes['XOM']
    x = sm.add_constant(x)
    model = sm.OLS(y, x)
    results = model.fit()
    rsquared_data[column] = results.rsquared
```

```python
pd.Series(rsquared_data).sort_values(ascending=False)
```
