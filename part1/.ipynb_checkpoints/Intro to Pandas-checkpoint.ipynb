# Pandas Challenge #2

## 1. Which geography has the highest median income?

```python
df.sort_values(by='Median HHIncome', ascending=False).head()
```

```python
df[df['Median HHIncome'] == df['Median HHIncome'].max()]
```

## 2. What large geography (1MM+) has the smallest average family size?

```python
df[df['Population'] > 1000000].sort_values(by='Average family size').head()
```

```python
df[df['Average family size'] == df['Average family size'].min()]
```

```python
large_cities_df = df[df['Population'] > 1000000].copy()
large_cities_df[large_cities_df['Average family size'] == large_cities_df['Average family size'].min()]
```

## 	3.	Which geographies in NY have the highest and lowest median age?

```python
df_ny = df[df['Geography'].str.contains("NY")]
df_ny.sort_values(by='Median Age').head()
```

```python
df_ny.sort_values(by='Median Age', ascending=False).head()
```

```python
df_ny[df_ny['Median Age'] == df_ny['Median Age'].min()]
```

```python
df_ny[df_ny['Median Age'] == df_ny['Median Age'].max()]
```

```python
df_ny[(df_ny['Median Age'] == df_ny['Median Age'].max()) | (df_ny['Median Age'] == df_ny['Median Age'].min())]
```

# Statistics with StatsModels

```python
import statsmodels.api as sm
```

```python
x = df['Population']
y = df['Total households']
```

```python
x = sm.add_constant(x)
```

```python
model = sm.OLS(y, x)
results = model.fit()
results.summary()
```

```python
results.rsquared
```

```python
results.params
```

```python
results.tvalues
```

```python
results.pvalues
```
